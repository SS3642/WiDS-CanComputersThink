```python
import numpy as np
import random

class Agent:
    def __init__(self):
        self.q = {}
        self.l = 0.1
        self.d = 0.9
        self.e = 0.1

    def gq(self, s):
        k = str(s)
        if k not in self.q:
            self.q[k] = [0.0] * 4
        return self.q[k]

    def act(self, s):
        if random.random() < self.e:
            return random.randint(0, 3)
        return np.argmax(self.gq(s))

    def update(self, s, a, r, ns):
        v = self.gq(s)
        nv = self.gq(ns)
        v[a] = v[a] + self.l * (r + self.d * max(nv) - v[a])

def step(s, a):
    p = np.where(s == 0)[0][0]
    r, c = p // 4, p % 4
    nr, nc = r, c
    if a == 0 and r > 0: nr -= 1
    elif a == 1 and r < 3: nr += 1
    elif a == 2 and c > 0: nc -= 1
    elif a == 3 and c < 3: nc += 1
    ns = s.copy()
    np_idx = nr * 4 + nc
    ns[p], ns[np_idx] = ns[np_idx], ns[p]
    return ns

# Main Training Loop
goal = np.append(np.arange(1, 16), 0)
ag = Agent()

for i in range(1000):
    s = goal.copy()
    random.shuffle(s) # Start with shuffled board
    for _ in range(100):
        a = ag.act(s)
        ns = step(s, a)
        r = 100 if np.array_equal(ns, goal) else -1
        ag.update(s, a, r, ns)
        s = ns
        if r == 100: break
```
